### Task 1 - Predict the percentage of an student based on the no. of study hours.

In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

### Importing the required libraries

```python
# Importing all libraries required in this notebook
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt  
```

```python
%matplotlib inline
print('Libraries imported successfully')
```

### Reading and Imprting Data from Online Source 

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)
print("Data imported successfully")

s_data.head(10)
```

### Data Visualization and Plotting The Distribution of Scores 

```python
# visualizing with line plot 
plt.style.use('ggplot')
s_data.plot(kind='line')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# visualizing with scatter plot
s_data.plot(kind='scatter',x='Hours',y='Scores',color='b',figsize=(8,5))
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
By visualization, we come to know that this problem can be easily solved by linear regression.
```

### Preparing the Data 

```python
x=np.asanyarray(s_data[['Hours']]) 
y=np.asanyarray(s_data['Scores'])
```

```python
# using the train test split to split the data to train and test the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
print(x_test)
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()  
regressor.fit(x_train, y_train) 

print("Training complete. \n")
```

```python
print('Coefficient: ', regressor.coef_)
print('intercept: ', regressor.intercept_)
```

```python
# plotting the regression line over the test data in simple linear regression.
s_data.plot(kind='scatter',x='Hours', y='Scores', figsize=(8,5), color='b') 
plt.plot(x_train, regressor.coef_[0]*x_train + regressor.intercept_, color='r')

plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
The red line is the best fit line for this data.
```

### Assessment of the model 

```python
# using metrics to find the mean absolute error and r2 to find the accuracy
from sklearn import metrics

from sklearn.metrics import r2_score

pred_y=regressor.predict(x_test)

print('Mean Absolute Error: {}'.format(metrics.mean_absolute_error( y_test, pred_y))) 
print("R2-score: %.2f" % r2_score( y_test, pred_y))
```

Mean Absolute Error : mean of absolute value of errors.
R2-score : metric for the accuracy of the model. Higher the R2 value higher is the accuracy of the model. Best is 1

```python
# Comparing Actual vs Predicted
df2 = pd.DataFrame({'Actual': y_test, 'Predicted': pred_y})
df2
```
